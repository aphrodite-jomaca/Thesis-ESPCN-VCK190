```python
# Custom Im2Col
def img2col(input, weight, padding=(0,0), stride=(1,1), dilation=(1,1), bias=None):
    batch_size, in_channels, in_h, in_w = input.shape
    out_channels, in_channels, kh, kw =  weight.shape

    unfold = torch.nn.Unfold(kernel_size=(kh, kw), dilation=dilation, padding=padding, stride=stride)

    inp_unf = unfold(input)

    out_h = (in_h + 2 * padding[0] - (kh - 1) - 1) / stride[0] + 1
    out_w = (in_w + 2 * padding[1] - (kw - 1) - 1) / stride[1] + 1
    out_h, out_w = int(out_h), int(out_w)

    w_r = weight.view(weight.size(0), -1).t()
    pad_w = (0, 0, 7, 0)

    wpad = F.pad(w_r, pad_w)

    pad = (0, 0, 7, 0)

    pad_img = F.pad(inp_unf, pad)
    print(wpad.shape)
    print(pad_img.shape)
    if bias is None:
        out_unf = pad_img.transpose(1, 2).matmul(wpad).transpose(1, 2)
    else:
        out_unf = (pad_img.transpose(1, 2).matmul(wpad) + bias).transpose(1, 2)
    print(out_unf.shape)
    out = out_unf.view(batch_size, out_channels, out_h, out_w)

    return pad_img, wpad, out_h, out_w, out
```
